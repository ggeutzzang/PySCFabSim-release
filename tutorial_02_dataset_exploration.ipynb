```python
import pandas as pd
import os

# Îç∞Ïù¥ÌÑ∞ÏÖã Í≤ΩÎ°ú
dataset_path = 'datasets/SMT2020_HVLM/'
```

## üìÅ 1. Ïñ¥Îñ§ ÌååÏùºÎì§Ïù¥ ÏûàÎÇòÏöî?

```python
# Îç∞Ïù¥ÌÑ∞ÏÖã ÎîîÎ†âÌÜ†Î¶¨Ïùò Î™®Îì† ÌååÏùº Î™©Î°ù
files = os.listdir(dataset_path)
print("Îç∞Ïù¥ÌÑ∞ÏÖã ÌååÏùº Î™©Î°ù:")
for f in sorted(files):
    file_size = os.path.getsize(os.path.join(dataset_path, f))
    print(f"  - {f:20s} ({file_size:,} bytes)")
```

## üè≠ 2. tool.txt.1l - Î®∏Ïã†(Ïû•ÎπÑ) Ï†ïÎ≥¥

Ïñ¥Îñ§ Î®∏Ïã†Îì§Ïù¥ ÏûàÎäîÏßÄ ÏÇ¥Ìé¥Î¥ÖÏãúÎã§.

```python
# tool.txt.1l ÌååÏùº ÏùΩÍ∏∞ (ÌÉ≠ÏúºÎ°ú Íµ¨Î∂Ñ)
df_tool = pd.read_csv(dataset_path + 'tool.txt.1l', sep='\t')

print(f"Ï¥ù {len(df_tool)}Í∞úÏùò Î®∏Ïã† ÌÉÄÏûÖÏù¥ ÏûàÏäµÎãàÎã§.\n")
print("Ï≤òÏùå 10Í∞ú Ìñâ:")
df_tool.head(10)
```

```python
# Ï£ºÏöî Ïª¨ÎüºÎßå Î≥¥Í∏∞
df_tool[['STNFAM', 'STNQTY', 'STNGRP', 'STNFAMLOC']].head(10)
```

```python
# Ï¥ù Î®∏Ïã† Ïàò Í≥ÑÏÇ∞
total_machines = df_tool['STNQTY'].sum()
print(f"\nÏ†ÑÏ≤¥ Î®∏Ïã† Ïàò: {total_machines:.0f}ÎåÄ")

# Í∑∏Î£πÎ≥Ñ Î®∏Ïã† Ïàò
print("\nÍ∑∏Î£πÎ≥Ñ Î®∏Ïã† Ïàò:")
df_tool.groupby('STNGRP')['STNQTY'].sum().sort_values(ascending=False)
```

## üìã 3. order.txt - Ï£ºÎ¨∏ Ï†ïÎ≥¥

Ïñ¥Îñ§ Ï†úÌíàÏùÑ ÏñºÎßàÎÇò ÎßåÎì§Ïñ¥Ïïº ÌïòÎäîÏßÄ Î¥ÖÏãúÎã§.

```python
# order.txt ÌååÏùº ÏùΩÍ∏∞
df_order = pd.read_csv(dataset_path + 'order.txt', sep='\t')

print(f"Ï¥ù {len(df_order)}Í∞úÏùò Ï£ºÎ¨∏Ïù¥ ÏûàÏäµÎãàÎã§.\n")
df_order.head(10)
```

```python
# Ï£ºÏöî Ï†ïÎ≥¥Îßå Î≥¥Í∏∞
df_order[['PART', 'PRIOR', 'RPT#', 'PIECES', 'START', 'DUE']].head(10)
```

```python
# Ïö∞ÏÑ†ÏàúÏúÑÎ≥Ñ Ï£ºÎ¨∏ Ïàò
print("Ïö∞ÏÑ†ÏàúÏúÑÎ≥Ñ Ï£ºÎ¨∏ Í∞úÏàò:")
df_order['PRIOR'].value_counts().sort_index()
```

```python
# Ï†úÌíàÎ≥Ñ Ï¥ù ÏÉùÏÇ∞ ÏàòÎüâ (RPT# * PIECES)
df_order['total_pieces'] = df_order['RPT#'] * df_order['PIECES']

print("\nÏ†úÌíàÎ≥Ñ Ï¥ù ÏÉùÏÇ∞ ÏàòÎüâ:")
df_order.groupby('PART')['total_pieces'].sum().sort_values(ascending=False)
```
